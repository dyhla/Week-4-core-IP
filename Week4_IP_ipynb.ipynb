**Linear correlation**

```python

# Calculating the pearson coefficient
pearson_coeff = score["Utilib_14_taken_sum"].corr(score["BlueCars_taken_sum"], method="pearson") 
print(pearson_coeff)

# Checking whether you have to define the pearson
coeff = score["Utilib_14_taken_sum"].corr(score["BlueCars_taken_sum"]) 
print(coeff)
```

There is a positive correlation between the variables
